```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Titanic-cleaned.csv")
```

```python
df.head(10)
```

```python
#Feature Interaction

df['family_size'] = df['SibSp']+df['Parch']+1
df["fare_per_person"] = df["Fare"] / df["family_size"]
```

```python
df.head(10)
```

```python
#Aggregation Features to calculate average mean

df["avg_age_class"] = df.groupby("Pclass")["Age"].transform("mean")

```

```python
df["avg_age_class"].head(10).round()
```

```python
#average fare price 
df["avg_fare_class"] = df.groupby("Pclass")["Fare"].transform("mean")
```

```python
df["avg_fare_class"].head(10).round()
```

```python
#Age group feature

df["age_group"] = pd.cut(df["Age"],bins=[0,12,18,60,100],labels=["Child","Teen","Adult","Senior"])
```

```python
df["age_group"].head(10) 
```

```python
#retail sales dataset to perform rolling statistics

df = pd.read_csv("retail_sales_dataset.csv")
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Date'].head()
```

```python
df = df.sort_values("Date")
```

```python
df["year"] = df["Date"].dt.year
df["month"] = df["Date"].dt.month
df["day"] = df["Date"].dt.day
df["weekday"] = df["Date"].dt.weekday
```

```python
df.head(10)
```

```python
df["rolling_7day_sales"] = df["Total Amount"].rolling(7).mean()
```

```python
df["rolling_max_sales"] = df["Total Amount"].rolling(7).max()
```

```python
df["rolling_sales_std"] = df["Total Amount"].rolling(7).std()
```
