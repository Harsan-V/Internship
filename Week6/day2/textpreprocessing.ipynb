```python
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)

data = response.json()
df = pd.DataFrame(data)

print(df.head())
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    text = clean_text(text)
    
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words]
    tokens = [lemmatizer.lemmatize(w, pos='v') for w in tokens]
    
    return tokens
```

```python
df['processed'] = df['body'].apply(preprocess)

print(df[['body', 'processed']].head())
```

```python
from collections import Counter

all_words = []

for tokens in df['processed']:
    all_words.extend(tokens)

word_freq = Counter(all_words)

print(word_freq.most_common(10))
```

```python
texts = [
    "I LOVED the phone!!! Battery-life is amazing :-) Visit https://shop.example.com",
    "Worst delivery ever... package was delayed by 5 days &amp; support was slow.",
    "@brand Great camera, but the apps are crashing again and again.",
    "The shoes are running small, though the design looks beautiful!",
    "Returned it. The product was broken, dirty, and missing parts.",
]


df = pd.DataFrame({"raw_text": texts})
df
```

```python
punct_translator = str.maketrans("", "", string.punctuation)


def clean_text(text):
    text = html.unescape(text)                         # Convert &amp; to &
    text = text.lower()                                # Lowercase
    text = re.sub(r"https?://\S+|www\.\S+", " ", text) # Remove URLs
    text = re.sub(r"@\w+|#\w+", " ", text)              # Remove mentions/hashtags
    text = text.replace("-", " ")                       # Split hyphenated words
    text = text.translate(punct_translator)            # Remove punctuation
    text = re.sub(r"\d+", " ", text)                     # Remove numbers
    text = re.sub(r"\s+", " ", text).strip()             # Remove extra spaces
    return text


df["clean_text"] = df["raw_text"].apply(clean_text)
df[["raw_text", "clean_text"]]
```

```python
def tokenize(text):
    return re.findall(r"[a-z]+", text)


df["tokens"] = df["clean_text"].apply(tokenize)
df[["clean_text", "tokens"]]
```

```python
stopwords = {
    "a", "am", "and", "are", "but", "by", "for", "i", "is", "it",
    "of", "the", "this", "though", "to", "was", "were"
}


def remove_stopwords(tokens):
    return [word for word in tokens if word not in stopwords]


df["tokens_no_stopwords"] = df["tokens"].apply(remove_stopwords)
df[["tokens", "tokens_no_stopwords"]]
```

```python
lemmatizer = WordNetLemmatizer()


def guess_wordnet_pos(word):
    if word.endswith("ing") or word.endswith("ed"):
        return "v"
    if word.endswith("ly"):
        return "r"
    if word.endswith("ful") or word.endswith("ous") or word.endswith("ive"):
        return "a"
    return "n"


def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(word, guess_wordnet_pos(word)) for word in tokens]


df["lemmas"] = df["tokens_no_stopwords"].apply(lemmatize_tokens)
df[["tokens_no_stopwords", "lemmas"]]
```

```python
stemmer = PorterStemmer()


def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]


df["stems"] = df["tokens_no_stopwords"].apply(stem_tokens)
df[["tokens_no_stopwords", "lemmas", "stems"]]
```

```python
def preprocessing_pipeline(text):
    cleaned = clean_text(text)
    tokens = tokenize(cleaned)
    tokens = remove_stopwords(tokens)
    lemmas = lemmatize_tokens(tokens)
    stems = stem_tokens(tokens)
    return {
        "cleaned": cleaned,
        "tokens": tokens,
        "lemmas": lemmas,
        "stems": stems,
    }


sample = "The customers were happily running and recommending the beautiful products!"
preprocessing_pipeline(sample)
```

```python
df["lemma_text"] = df["lemmas"].apply(" ".join)
df["stem_text"] = df["stems"].apply(" ".join)

df[["raw_text", "clean_text", "tokens_no_stopwords", "lemma_text", "stem_text"]]
```
