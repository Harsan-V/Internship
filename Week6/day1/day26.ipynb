```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

sentence = "Hi my name is Harshith. I am a Computer Science student."

words = word_tokenize(sentence)

filtered = [w for w in words if w.lower() not in stopwords.words('english')]

print(filtered)
```

NLP usecase - Sentiment Analysis

```python
from sklearn.naive_bayes import MultinomialNB

texts = ["good product", "bad experience", "excellent quality", "worst item"]
labels = [1, 0, 1, 0]  # 1 = positive, 0 = negative

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

model = MultinomialNB()
model.fit(X, labels)
```

```python
test = ["good quality", "bad product"]
X_test = vectorizer.transform(test)

predictions = model.predict(X_test)
print(predictions)
```
